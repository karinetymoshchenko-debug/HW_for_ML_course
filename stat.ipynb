Вибіркове середнє (mean) доречне тоді, коли:
дані симетричні або близькі до нормального розподілу
немає викидів (дуже великих або дуже малих значень)
важливо врахувати внесок кожного значення
Медіана краща, коли:
дані несиметричні, “косі”
є викиди, які сильно тягнуть середнє
важливо знайти “центральне” значення, яке не спотворюється екстремами
розподіл має довгий хвіст

Для оцінки середнього чека використала б середнє вибіркове, щоб врахувати всі значення вибірки. А для прогнозів оборотів (продажів) - медіану, щоб вірогідність була точніша.
