### Домашні завдання на відпрацювання поняття функції та її похідної

1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
x = sp.symbols('x')
f_lin = 10*x - 5
derivative_f_lin = sp.diff(f_lin, x)
print("Похідна лінійної функції 10x-5:", derivative_f_lin)
```

```python
# Дані
x = np.linspace(0, 10, 200)
f_lin = 10*x - 5
derivative_f_lin = np.full_like(x, 10)

# Точка та дотична
x_point = 3
f_point = 10*x_point - 5
df_point = 10
tangent_line = df_point * (x - x_point) + f_point

# Графік
plt.figure(figsize=(10, 5))
plt.plot(x, f_lin, label='f(x) = 10x - 5', color='blue')
plt.plot(x, derivative_f_lin, label="f'(x) = 10", color='green')
plt.scatter(x_point, f_point, color='red')
plt.plot(x, tangent_line, '--', color='red', label='Tangent at x = 3')

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')
f_sin = sp.sin(x)
derivative_f_sin = sp.diff(f_sin, x)
print("Похідна  функції sin(x):", derivative_f_sin)
```

```python
x = np.linspace(0, 5, 10)
f_sin = np.sin(x)
derivative_f_sin = np.cos(x)

# Точка та дотична
x_point = 1
f_point = np.sin(x_point)
df_point = np.cos(x_point)
tangent_line = df_point * (x - x_point) + f_point

# Графік
plt.figure(figsize=(10, 5))
plt.plot(x, f_sin, label='f(x) = sin(x)', color='blue')
plt.plot(x, derivative_f_sin, label="f'(x) = cos(x)", color='green')
plt.scatter(x_point, f_point, color='red')
plt.plot(x, tangent_line, '--', color='red', label='Tangent at x = 1')

plt.title("Графік функції sin(x)  та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = sp.symbols('x')
f_ln = sp.ln(x)
derivative_f_ln = sp.diff(f_ln, x)
print("Похідна  функції ln(x):", derivative_f_ln)
```

```python
x = np.linspace(0.1, 5, 10)
f_ln = np.log(x)
derivative_f_ln = 1/x

# Точка та дотична
x_point = 3
f_point = np.log(x_point)
df_point = 1/(x_point)
tangent_line = df_point * (x - x_point) + f_point

# Графік
plt.figure(figsize=(10, 5))
plt.plot(x, f_ln, label='f(x) = log(x)', color='blue')
plt.plot(x, derivative_f_ln, label="f'(x) = 1/x", color='green')
plt.scatter(x_point, f_point, color='red')
plt.plot(x, tangent_line, '--', color='red', label='Tangent at x = 3')

plt.title("Графік функції log(x)  та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
f_par = x**2 - 5*x + 1
derivative_f_par = sp.diff(f_par, x)
print("Похідна  функції x**2 - 5*x + 1:", derivative_f_par)
```

```python
x = np.linspace (0, 100, 100)
f_par = x**2 - 5*x + 1
derivative_f_par = 2*x - 5

# Точка та дотична
x_point = 40
f_point = (x_point)**2 - 5*(x_point) + 1
df_point = 2*(x_point)-5
tangent_line = df_point * (x - x_point) + f_point

# Графік
plt.figure(figsize=(10, 5))
plt.plot(x, f_par, label='f(x) = x**2 - 5*x + 1', color='blue')
plt.plot(x, derivative_f_par, label="f'(x) = 2*x - 5", color='green')
plt.scatter(x_point, f_point, color='red')
plt.plot(x, tangent_line, '--', color='red', label='Tangent at x = 40')

plt.title("Графік функції x**2 - 5*x + 1  та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
