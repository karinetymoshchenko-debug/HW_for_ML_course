# Цикл for - базовий

```python
for i in range (1,8):
   print(i)
```

# Список

```python
names = ["Олег", "Анна", "Іван", "Марія"]
for name in names:
  print(f"Привіт, {name}!")
```

# Цикл з enumerate

```python
grades = [85, 92, 78, 95, 88]
for i, grade in enumerate(grades):
  print(f"Студент {i+1}: {grade} балів")
```

# Сума парних чисел

```python
#Option1
numbers=list (range (0,21,2))
print(numbers)
print(sum(numbers))
```

```python
#Option2
numbers=list(range (0,21,1))
total=0
for num in numbers:
    if num % 2 == 0:
        total=total+num
print(numbers)
print(total)

```
