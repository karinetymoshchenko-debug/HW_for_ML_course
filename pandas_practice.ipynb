5. Статистика із групуванням

Обчисліть (кожен пункт оцінюється в 0.5 бала):

Середній рахунок за статтю (groupby sex)
Середні чайові за днем тижня (groupby day)
Скільки відвідувачів у кожен день

```python
avg_bill_by_sex = df.groupby('sex')['total_bill'].mean() #Середній рахунок за статтю (groupby sex)
print('Середній чек за статтю')
avg_bill_by_sex
```

```python
avg_tip_by_day = df.groupby('day')['tip'].mean() #Середні чайові за днем тижня (groupby day)
print ('Середні чайові по днях')
avg_tip_by_day
```

```python
visitors_by_day = df['day'].value_counts() #Скільки відвідувачів у кожен день
print ('Кількість відвідувачів по днях тижня')
visitors_by_day
```
