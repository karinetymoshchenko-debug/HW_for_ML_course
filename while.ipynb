# 1. Цикл while

```python
counter = -3
while counter < 6:
    counter += 1
    print(f"Лічильник {counter}")
```

# 2. List comprehension - прості операції

```python
numbers = [11, 4, 9, 16, 7, 5]

squares = [num**2 for num in numbers]#Squares
print (squares)

for num in numbers: #парні числа
  if num % 2 == 0:
    print(num)

new_num=[num*10 for num in numbers] #множення на 10
print(new_num)
```

# 3.List comprehension - комбінована логіка

```python
numbers = list(range(1, 11))
print(numbers)
```

```python
numbers = list(range(1, 11))
[x*2 if x % 2 == 0 else x-1 for x in numbers]
```
