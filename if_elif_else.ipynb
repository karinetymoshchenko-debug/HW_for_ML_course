# if-elif-else

```python
cooking_time = 5
if cooking_time < 4:
  print ("Яйце рідке")
elif 4 <= cooking_time < 7:
  print("Яйце всмятку")
elif 7<= cooking_time <10:
  print("Яйце вкруту")
else:
  print("Яйце переварене")
```
