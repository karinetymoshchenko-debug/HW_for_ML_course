**1. Функція з параметром**

Створіть функцію `show_menu()`, яка приймає перелік страв, наприклад ['Піца', 'Паста', 'Салат'] і виводить меню ресторану в наступному вигляді:
```
=== МЕНЮ ===
1. Піца
2. Паста
3. Салат
=============
```

Викличте цю функцію на списку ["Суп грибний", "Котлета по-київськи", "Лимонад"].

````python
def show_menu(dishes):
    """
  This function asksfor menu.
  Specify dishes.

  params:
  show_menu: str
  """
    print ("===МЕНЮ===")
    for i, dish in enumerate(dishes, start=1):
        print (f"{i}.{dish}")
    print("=============")
    
````

````python
show_menu(['Піца', 'Паста', 'Салат'])
````

````python
show_menu(["Суп грибний", "Котлета по-київськи", "Лимонад"])
````

````python
?show_menu
````

**2. Функція з параметром**

Створіть функцію `print_square(n)`, яка виводить квадрат числа.

Наприклад: `print_square(5)` → "Квадрат числа 5 = 25"

Викличте її з різними числами.

````python
def print_square(n):
    """
    This function calcs squre of a number.
    Specify number.

    params:
    print_square: int
    """    
    square=n**2
    print (f"Квадрат числа {n} = {square}")
````

````python
print_square(5)
````

````python
print_square(-10)
````

````python
print_square(0)
````

````python
print_square() #додаткові перевірки
````

````python
print_square("cat") #додаткові перевірки
````


**4. Функція пошуку максимумального значення**

Створіть функцію `find_max(numbers)`, яка:
- Приймає список чисел
- Повертає найбільше число в списку

Протестуйте на списку `[15, 42, 8, 23, 56, 19]`.


````python
def find_max(numbers):
    """
    This function is searching for maximum in a integer list.
    Specify  integer list.

    params:
    find_max:  integer list
    """
    return max(numbers)
  
````

````python
find_max([15, 42, 8, 23, 56, 19])
````

**5. Функція з умовами**

Створіть функцію `check_grade(grade)`, яка:
- Приймає оцінку (число)
- Повертає "Здав" якщо grade >= 60
- Повертає "Не здав" якщо grade < 60

Протестуйте на різних оцінках.


````python
def check_grade(grade):
    """
    This function check grade and print result.
    Specify  int.

    params:
    chek_grade: int
    """
    if  60 <= grade <=100:
        print ("Здав")
    if 1 <= grade < 60:
        print ("Не здав")
    if grade < 1:
        print ("Перевірте коректність введеного значення")
    if grade > 100:
        print ("Перевірте коректність введеного значення")
````

````python
check_grade(90)
````

````python
check_grade(59)
````

````python
check_grade(60)
````
