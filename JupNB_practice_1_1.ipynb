```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 11)    
y = 2**x            
```

```python
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Графік функції y = 2^x")
plt.grid(True)
```
